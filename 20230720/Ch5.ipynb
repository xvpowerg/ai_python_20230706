```python
fruits = ["香蕉","蘋果","橘子","西瓜"]
while True:
    fruit = input("請輸入水果(Enter結束)")
    if fruit == "":
        break
    n = fruits.count(fruit)
    if n == 0:
        print(f"{fruit}不在清單內")
    else:
        n = fruits.index(fruit)
        print(f"{fruit}在清單內在第{n}")
        
```

## 輸入四個月的支出金額後列出最多與最少的支出金額

```python
priceList=[]
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額"))
    priceSum =priceSum + p
    priceList.append(p)
priceList.sort()
print(f"max:{priceList[len(priceList)-1]}") 
print(f"min:{priceList[0]}")
print(priceSum)    
print(priceList) 

print(sum(priceList)) 
print(max(priceList))    
print(min(priceList)) 
```

## 匿名函數
不需要定義函數名稱，只需要用到運算式或表達分析語法。

Python使用lambda語法定義匿名函數。

匿名函數是一個表達式或計算式，並不是一個執行流程區塊。

匿名函數出現在一般函數不允許的地方，例如list裡面或著函數呼叫參數的位置。


```python
def f(x,y,z):
    return x+y+z
print(f(2,30,400))

f2 = lambda x,y,z:x+y+z
print(f2(2,30,400))
```

```python
def f1(x):
    return x ** 2
def f2(x):
    return x ** 3
def f3(x):
    return x ** 4
list1 = [f1,f2,f3]
for f in list1:
    print(f(3))
```

```python
list2 = [lambda x: x**2,lambda x : x**3,lambda x: x**4]
for f in list2:
    print(f(3))
    
```

## Map

```python
list3 =[1,2,3,'',4,5,'']
mapObj = map(lambda x : x if x != '' else 0,list3) #把空白轉換成0
print(list(mapObj))
```

## Filter

```python
a = [1,2,3,4,5,6]
filterObj = filter(lambda x : x % 2 == 0,a)
print(list(filterObj))
```

```python
list3 =[1,2,3,'',4,5,'']
#不是''幫我輸出成list
filterObj = filter(lambda x : x != '',list3)
print(list(filterObj))
```

## 例外語法
try:
    保護的程式區段
    
except 例外1 :
    例外1發生時執行的程式
    
except :
    其他例外發生時執行的程式
    #未指定例外型態，捕捉所有例外物件
    
else:
    若try程式區段沒產生例外執行的程式
    
finally:
    不管有沒有發生例外都會執行的程式

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    print(num1/num2)
except:
    print("有錯誤發生")
    
print("End")
```

```python
# 詳細分類錯誤類型
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    #print("A")
    #print(num1/num2)
    nums[9]
    #int("A")
except ZeroDivisionError:
    print("Error發生分母為0")
except IndexError:
    print("Error發生Index錯誤")
except:
    print("不在上面分類的例外")
else:
    print("沒有發生例外") #沒有任何例外發生時執行
finally:
    print("一定會執行的區塊") #釋放資源
print("End")
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = 'one, two...ten'
re1 = re.split("[,. ]+",text)
re2 = re.split("[,. ]+",text,maxsplit=1)#只分割一次
print(re1)
print(re2)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456"
rStr = re.sub("\d+","222",inputStr)
print(rStr)
rStr = re.sub("\d+","222",inputStr,count=1)
print(rStr)
```

## 繼承

```python
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def printInfo(self):    
        print(self.name,self.age)     
an = Animal("Bobo",5)        
an.printInfo()
```
