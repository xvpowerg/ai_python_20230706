```python
import pandas as pd
fortune = pd.read_csv("fortune1000.csv",index_col='Rank')
print(fortune)
print("="*100)
sectors=fortune.groupby("Sector")# 部門
print(sectors.size())
```

```python
print(sectors.get_group('Energy').head(5))
print("="*100)
print(sectors.get_group('Energy').max()) #找出所有欄位的最大值 是能源部門
print("="*100)
print(sectors.get_group('Energy').min())#找出所有欄位的最小值 是能源部門
```

```python
print(sectors.sum())
print()
print(sectors.mean())
print()
```

```python
print(sectors.agg("sum"))
print(sectors.agg(["size","sum","mean"]))
print("=======================================")
print(sectors.agg({"Revenue":["size","sum","mean"],"Employees":"mean"}))
```

```python
import pandas as pd
auto_price=pd.read_csv('Automobile.csv') #看一下Automobile.csv 哪些是?
print(auto_price.head())
print('了解欄位名稱，-容易有被誤判的風險')
print(auto_price.columns)
print('儲存欄位名稱')
```

```python
print('查看資料內是否有遺失值')
print(auto_price.isnull().any())
print('查看每個欄位的資料型態')
print(auto_price.dtypes)
```

```python
import numpy as np
print('查看哪些欄位有？')
list1=[]
for col in auto_price:
    for filed1 in auto_price[col]:
          if filed1=='?':
                list1.append(col)
print(list1)
print(len(list1))
print('太多資料重複，希望資料不要重複出現，我們將資料轉換為set型態')
set1=set(list1)
print(set1)
print(len(set1))
list2=list(set1)
```

```python
for column in list2:
    auto_price.loc[auto_price[column] == "?",column] = np.NaN
print(auto_price.isnull().any())
print(auto_price.dtypes)
```

```python
print('刪除之前資料的形狀')
print(auto_price.shape)
#print(type(auto_price.isnull()) )
print(auto_price.isnull().sum())
try:
    auto_price.drop(["normalized-losses"],axis=1,inplace = True)
except:
    pass
print(auto_price.shape)
```

```python
print(auto_price.isnull().sum())
auto_price.dropna(axis=0,inplace=True)
print(auto_price.shape)
```

```python
list3 = []
for i in auto_price.columns:
    
    if auto_price[i].dtype == object:
        try:
            auto_price[i] = auto_price[i].astype('float')
            list3.append(i)
        except:
            pass
print("轉換為float的欄位:")        
print(list3)        
```

```python
print('描述數值型態資料')
print(auto_price.describe( ))
print('描述字串型態資料')
print(auto_price.describe(include=['object']))
```

```python
print('分別挑選出數值型態與字串型態欄位')
cols4=[]#字串
cols5=[]#數值
for i in auto_price.columns:
  if auto_price[i].dtype==object:
    cols4.append(i)
  else:
    cols5.append(i)
print('字串欄位有：\n',cols4)
print('數值欄位有：\n',cols5)
print('字串型資料內資料出現的筆數統計')
for column in cols4:
  print('欄位為：',column)
  print(auto_price[column].value_counts( ))
  print( )
```

```python
print('字串資料以長條圖顯示')
import matplotlib.pyplot as plt
for cols in cols4:
  auto_price[cols].value_counts( ).plot(
      kind='bar',title=cols,rot=45
  )
  plt.show( )
```

```python
print('數值資料與price之間關係，以散佈圖顯示')
for cols in cols5:
  auto_price.plot(
      kind='scatter',
      title=cols,x=cols,y='price',rot=45
  )
  plt.show( )
```

```python
print('想要了解多個欄位之間的關係')
import seaborn as sns
sns.heatmap(auto_price[cols5].corr( ))
plt.show( )
```

```python
print('數值資料以直方圖顯示')
import matplotlib.pyplot as plt
for cols in cols5:
  auto_price[cols].plot(
      kind='hist',bins=12,title=cols,rot=45
  )
  plt.show( )
```
